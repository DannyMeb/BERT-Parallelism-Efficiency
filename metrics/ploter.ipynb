```python
# Load and display the entire content of the JSON file to manually verify the keys and data
import json
file_path = 'baseline_metrics.json'

with open(file_path, 'r') as file:
    data = json.load(file)

# Display the keys and a sample of data to ensure we understand the structure and correct key usage
print("Keys available in JSON:", list(data.keys()))
print("Sample data for 'statistical_efficiency':", data['statistical_efficiency'][:10])
print("Sample data for 'gpu_utilization':", data['gpu_utilizations'][:10])
print("Sample data for 'throughputs':", data['throughputs'][:10])
print("Sample data for 'losses':", data['losses'][:10])
```

```python
import matplotlib.pyplot as plt

# Extract each metric data
statistical_efficiency = data['statistical_efficiency']
gpu_utilizations = data['gpu_utilizations']
throughputs = data['throughputs']
losses = data['losses']

# Create separate plots for each metric
# Statistical Efficiency
plt.figure(figsize=(10, 4))
plt.plot(statistical_efficiency, label='Statistical Efficiency', color='blue')
plt.title('Statistical Efficiency Over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Efficiency')
plt.legend()
plt.show()

# GPU Utilizations
plt.figure(figsize=(10, 4))
plt.plot(gpu_utilizations, label='GPU Utilization', color='orange')
plt.title('GPU Utilization Over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Utilization (%)')
plt.legend()
plt.show()

# Throughputs
plt.figure(figsize=(10, 4))
plt.plot(throughputs, label='Throughputs', color='green')
plt.title('Throughputs Over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Throughput (items/sec)')
plt.legend()
plt.show()

# Training Losses
plt.figure(figsize=(10, 4))
plt.plot(losses, label='Training Loss', color='red')
plt.title('Training Loss Over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
